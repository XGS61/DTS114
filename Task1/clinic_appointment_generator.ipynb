# DTS114 Task 1: AI-DLC Clinic Appointment Generator

This notebook generates an AI-DLC-informed Flask clinic appointment administration prototype.

Safety boundary: appointment administration only; no diagnosis, no treatment advice, no real patient records.

## Phase 1: Inception

The Inception phase defines the business problem, actors, requirements, user stories, and safety boundary before code is generated.

```python
from pathlib import Path
import json
from textwrap import dedent

cwd = Path.cwd()
submission_root = cwd.parent if cwd.name == 'Task1' else cwd
app_dir = submission_root / 'Task2' / 'clinic_app'
artifact_dir = app_dir / 'artifacts'
diagram_dir = artifact_dir / 'diagrams'
static_dir = app_dir / 'static'
for path in [app_dir, artifact_dir, diagram_dir, static_dir, app_dir / 'templates', app_dir / 'tests']:
    path.mkdir(parents=True, exist_ok=True)

business_problem = 'A small clinic needs a safe appointment administration prototype that collects requests, flags slot conflicts, and keeps receptionist review before confirmation.'
actors = ['Patient', 'Receptionist', 'Clinic System']
safety_boundary = {
    'scope': 'Appointment administration only',
    'not_allowed': ['diagnosis', 'treatment advice', 'prescriptions', 'real patient records'],
    'human_review': 'Receptionist confirms or cancels each request.'
}
print(business_problem)
print(safety_boundary)
```

```python
user_stories = [
    {
        'id': 1,
        'role': 'Patient',
        'goal': 'submit an appointment request',
        'benefit': 'the clinic can review my preferred time',
        'acceptance_criteria': [
            'Form collects display name, date, time, type, and optional administrative reason.',
            'Successful submission creates Pending Review status.',
            'Users are warned not to enter real patient records.'
        ]
    },
    {
        'id': 2,
        'role': 'Receptionist',
        'goal': 'confirm or cancel pending requests',
        'benefit': 'appointments stay under human control',
        'acceptance_criteria': [
            'Receptionist can set status to Confirmed.',
            'Receptionist can set status to Cancelled.',
            'Review note and timestamp are recorded.'
        ]
    },
    {
        'id': 3,
        'role': 'Clinic System',
        'goal': 'flag same-slot conflicts',
        'benefit': 'the receptionist can spot scheduling risk',
        'acceptance_criteria': [
            'Same date/time active requests set conflict=true.',
            'Conflicted requests remain Pending Review.',
            'Cancelled requests do not block future slots.'
        ]
    }
]
(artifact_dir / 'generated_user_stories.json').write_text(json.dumps({'user_stories': user_stories}, indent=2), encoding='utf-8')
print(json.dumps(user_stories, indent=2))
```

## Phase 2: Construction

The Construction phase generates SDLC documentation, UML source, a generated image, Flask API code, website code, and a validation checklist.

```python
requirements_md = f"""
# Generated Requirements

## Business Problem
{business_problem}

## Actors
{', '.join(actors)}

## Safety Boundary
- Scope: {safety_boundary['scope']}
- Not allowed: {', '.join(safety_boundary['not_allowed'])}
- Human review: {safety_boundary['human_review']}

## Core Requirements
1. Create appointment requests as Pending Review.
2. Flag same date/time conflicts.
3. Allow receptionist confirmation or cancellation.
4. Return non-diagnostic administrative summaries.
5. Provide tests, CI/CD, and deployment-ready configuration.
""".strip()
(artifact_dir / 'generated_requirements.md').write_text(requirements_md, encoding='utf-8')
print(requirements_md)
```

```python
use_case_puml = """@startuml
actor Patient
actor Receptionist
rectangle "Clinic Appointment Prototype" {
  usecase "Submit appointment request" as Submit
  usecase "Review pending request" as Review
  usecase "Confirm or cancel" as Decide
  usecase "Flag slot conflict" as Conflict
  usecase "Generate non-diagnostic summary" as Summary
}
Patient -- Submit
Receptionist -- Review
Review ..> Decide : <<include>>
Submit ..> Conflict : <<include>>
Review ..> Summary : <<include>>
@enduml
"""
(diagram_dir / 'generated_use_case_diagram.puml').write_text(use_case_puml, encoding='utf-8')
print(use_case_puml)
```
