## Construction: Generate Clinic Image

The generated website image is created locally with Pillow so the notebook remains reproducible without a paid image API.

````python
from PIL import Image, ImageDraw, ImageFont

image_path = static_dir / "generated_clinic_image.png"
width, height = 1000, 700
image = Image.new("RGB", (width, height), "#f4f7fb")
draw = ImageDraw.Draw(image)

try:
    title_font = ImageFont.truetype("arial.ttf", 54)
    label_font = ImageFont.truetype("arial.ttf", 28)
    small_font = ImageFont.truetype("arial.ttf", 22)
except OSError:
    title_font = ImageFont.load_default()
    label_font = ImageFont.load_default()
    small_font = ImageFont.load_default()

draw.rectangle((70, 80, 930, 620), fill="#ffffff", outline="#d9e2ec", width=4)
draw.text((120, 120), "Clinic Appointment System", fill="#17202a", font=title_font)
draw.rounded_rectangle((130, 230, 410, 410), radius=18, fill="#e8f5ef", outline="#17623f", width=3)
draw.text((170, 270), "Patient", fill="#17623f", font=label_font)
draw.text((170, 320), "Booking", fill="#17623f", font=label_font)
draw.rounded_rectangle((590, 230, 870, 410), radius=18, fill="#fff2dc", outline="#b4661f", width=3)
draw.text((635, 270), "Doctor/Admin", fill="#8a4f16", font=label_font)
draw.text((635, 320), "Review", fill="#8a4f16", font=label_font)
draw.line((415, 320, 585, 320), fill="#18646f", width=8)
draw.polygon([(585, 320), (555, 300), (555, 340)], fill="#18646f")
draw.rounded_rectangle((260, 470, 740, 555), radius=18, fill="#eef7f7", outline="#18646f", width=3)
draw.text((310, 500), "Shared Appointment Data", fill="#18646f", font=small_font)
image.save(image_path)
print(f"Generated image: {image_path}")
````
